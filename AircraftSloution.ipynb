## PartB 1

Steps involved
1.  Import city pair datasets
2.  Edit ac_char dataset and change column to type int
3.  Create a function to pick only those flights which has sufficient range
4.  Find out total no. of times a flight will have to do updown to meet the demand
    ---->A320 has capacity of 150 seats and range of 5000km
    ---->AA to BB we have demand of 420 and distance of 3000km
    ---->Clearly the flight has sufficient range..Now let's talk about capacity
    ---->Total passengers=420
    ---->>In one trip A320 will take 150 passengers from AA to BB (Total trip=1)
    ---->Flight will come back(Total trip=2)
    ---->In second trip will take anpother 150 passengers from AA to BB (Total trip=3)
    ---->Flight will come back(Total trip=4)
    ---->In third trip flight will take 120 passengers from AA to BB(Total trip=5)
    ---->In 5 updown flight A320 will meet the demand
5.  Pick only those flights which meet the demand in 3 or less trips
6.  Out the the flights in 5th Steps, pick the one with minimum price

```python
citypair=pd.read_csv("CityPair.csv")
```

```python
citypair["Distance"]=citypair["Distance "].str.replace(',', '').astype(int)
```

```python
citypair
```

```python
ac_char["Costs per flight hour"]=ac_char["Costs per flight hour"].str.replace(',', '').astype(int)
```

```python
ac_char["Range (Km)"]=ac_char["Range (Km)"].str.replace(',', '').astype(int)
```

```python
ac_char
```

```python
def func(value):
    a=list()
    for i,j in ac_char["Range (Km)"].iteritems():
        if j>value:
            a.append(i)
    return(a)
```

```python
size=len(citypair)
```

```python
citypair["AircraftSL"]=np.nan
```

```python
for i in range(size):
    citypair["AircraftSL"][i]=func(citypair["Distance"][i])
```

## These are the aircrafts which can fulfill the range requirement for each city pair

```python
 citypair["AircraftSL"]
```

```python
citypair
```

```python
citypair.drop("Distance ",axis=1,inplace=True)#Remove one Distance
```

```python
citypair
```

```python
import math
```

```python
def demand(value, aircraft):
    a=dict()
    for i,j in ac_char["Number of Seats"].iteritems():
        if i in aircraft:
            a[i]=(math.ceil(value/j))*2-1
    return(a)
```

Create a column of empty dictionary to keep the note of no. of trips each of the flight will take for each pair of cities

```python
citypair["ShortlistBySize"]= [{} for _ in range(len(citypair))]
```

```python
pd.set_option('display.max_colwidth', -1)# See complete rows columns without truncating
```

```python
for i in range(size):
    citypair["ShortlistBySize"][i]=demand(citypair["Demand"][i],citypair["AircraftSL"][i])
```

```python
citypair["ShortlistBySize"]
```

```python
citypair.rename(columns={"ShortlistBySize":"TotalUpDowntrips"},inplace=True)
```

```python
citypair
```

```python
def price(updown):
    a=dict()
    for i,j in ac_char["Costs per flight hour"].iteritems():
        if i in updown and updown[i]<=3:
            a[i]=j
    b=min(a, key=a.get)
    return(a,b)
```

```python
citypair["Cost"]=[{} for _ in range(len(citypair))] 
```

```python
citypair["MinimumCostFlight"]= np.nan #To figure out which flight has minimum cost
```

Cost of those flights which meet the demand in 3 or less trips

```python
for i in range(size):
    citypair["Cost"][i],citypair["MinimumCostFlight"][i]=price(citypair["TotalUpDowntrips"][i])
```

```python
citypair
```

## Clearly Minimum Cost Flight gives the Flight which has sufficient range, can manage the demands in 3 or less updown trips  and has minimum costs`

## So A330 and A220 are best suitable options
